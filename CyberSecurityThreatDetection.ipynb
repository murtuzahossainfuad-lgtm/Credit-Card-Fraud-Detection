```python
print("=" * 50)
print("CYBER SECURITY THREAT ANALYSIS 2015-2024")
print("=" * 50)

print("\n📊 DATASET INFO:")
print(f"Total Records: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"Time Period: 2015-2024")

print("\n🌍 COUNTRY ANALYSIS:")
print(f"Most Attacked: {df['Country'].value_counts().index[0]}")
print(f"Least Attacked: {df['Country'].value_counts().index[-1]}")

print("\n💣 ATTACK TYPE:")
print(f"Most Common: {df['Attack Type'].value_counts().index[0]}")
print(f"Least Common: {df['Attack Type'].value_counts().index[-1]}")

print("\n💰 FINANCIAL LOSS:")
most_loss = df.groupby('Attack Type')['Financial Loss (in Million $)'].mean().idxmax()
least_loss = df.groupby('Attack Type')['Financial Loss (in Million $)'].mean().idxmin()
print(f"Highest Loss Attack: {most_loss}")
print(f"Lowest Loss Attack: {least_loss}")

print("\n🏭 INDUSTRY:")
print(f"Most Targeted: {df['Target Industry'].value_counts().index[0]}")
print(f"Least Targeted: {df['Target Industry'].value_counts().index[-1]}")

print("\n🛡️ DEFENSE:")
best_defense = df.groupby('Defense Mechanism Used')['Incident Resolution Time (in Hours)'].mean().idxmin()
worst_defense = df.groupby('Defense Mechanism Used')['Incident Resolution Time (in Hours)'].mean().idxmax()
print(f"Most Effective: {best_defense}")
print(f"Least Effective: {worst_defense}")

print("\n⚠️ VULNERABILITY:")
print(f"Most Exploited: {df['Security Vulnerability Type'].value_counts().index[0]}")
print(f"Most Financial Loss: {df.groupby('Security Vulnerability Type')['Financial Loss (in Million $)'].mean().idxmax()}")

print("\n🎯 ATTACK SOURCE:")
print(f"Most Attacks: {df['Attack Source'].value_counts().index[0]}")
most_loss_source = df.groupby('Attack Source')['Financial Loss (in Million $)'].mean().idxmax()
print(f"Most Financial Loss: {most_loss_source}")

print("=" * 50)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Cyber Security Threat Analysis 2015-2024', fontsize=20, fontweight='bold')

# Chart 1 — Country
country_attacks = df['Country'].value_counts()
sns.barplot(ax=axes[0,0], x=country_attacks.index, y=country_attacks.values, palette='viridis', hue=country_attacks.index, legend=False)
axes[0,0].set_title('Attacks by Country', fontsize=13)
axes[0,0].set_xlabel('Country')
axes[0,0].set_ylabel('Count')
axes[0,0].tick_params(axis='x', rotation=45)

# Chart 2 — Attack Type
attack_counts = df['Attack Type'].value_counts()
sns.barplot(ax=axes[0,1], x=attack_counts.index, y=attack_counts.values, palette='magma', hue=attack_counts.index, legend=False)
axes[0,1].set_title('Attack Type Distribution', fontsize=13)
axes[0,1].set_xlabel('Attack Type')
axes[0,1].set_ylabel('Count')
axes[0,1].tick_params(axis='x', rotation=45)

# Chart 3 — Year Trend
yearly = df.groupby('Year')['Attack Type'].count()
sns.lineplot(ax=axes[0,2], x=yearly.index, y=yearly.values, marker='o', color='red', linewidth=2)
axes[0,2].set_title('Year-wise Trend', fontsize=13)
axes[0,2].set_xlabel('Year')
axes[0,2].set_ylabel('Count')

# Chart 4 — Industry
industry = df['Target Industry'].value_counts()
sns.barplot(ax=axes[1,0], x=industry.index, y=industry.values, palette='Set2', hue=industry.index, legend=False)
axes[1,0].set_title('Attacks by Industry', fontsize=13)
axes[1,0].set_xlabel('Industry')
axes[1,0].set_ylabel('Count')
axes[1,0].tick_params(axis='x', rotation=45)

# Chart 5 — Defense Effectiveness
defense = df.groupby('Defense Mechanism Used')['Incident Resolution Time (in Hours)'].mean().sort_values()
sns.barplot(ax=axes[1,1], x=defense.index, y=defense.values, palette='RdYlGn', hue=defense.index, legend=False)
axes[1,1].set_title('Defense Effectiveness', fontsize=13)
axes[1,1].set_xlabel('Defense')
axes[1,1].set_ylabel('Avg Resolution Time')
axes[1,1].tick_params(axis='x', rotation=45)

# Chart 6 — Vulnerability
vuln = df['Security Vulnerability Type'].value_counts()
sns.barplot(ax=axes[1,2], x=vuln.index, y=vuln.values, palette='rocket', hue=vuln.index, legend=False)
axes[1,2].set_title('Vulnerability Distribution', fontsize=13)
axes[1,2].set_xlabel('Vulnerability')
axes[1,2].set_ylabel('Count')
axes[1,2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
